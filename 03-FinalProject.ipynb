## Research Question

What is the relationship between household income and estimated residential carbon emission rates at the census tract level in San Diego County in 2022? With tract-level data, we measure income using Area Median Income (AMI) indicators as well as average annual household income. We can estimate residential carbon emissions using household electricity, natural gas, and other fuel expenditures. We consider this as a statistical inference problem to estimate the degree of association between income levels and emissions in order to understand whether or not income disparities are systematically associated with differences in residential carbon impact across San Diego County census tracts. 

## Background and Prior Work

Over the last fifty years, a growing number of eyes have been focused on the controversial topic of climate change, and by extension carbon emissions. Over this past half-decade, said emissions have skyrocketed dramatically<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1), prompting analysts to explore all aspects surrounding it -- ranging from its consequences, potential solutions, and---most relevant to this project---its roots. Due to this there exists many scientific journals and articles examining potential relationships between certain variables and carbon emissions, with one such variable being income level.

Discussions from scholars suggest an overall consensus that higher income levels, from local to national scales, correlate with higher levels of carbon emissions. Internationally, researcher scholars Sarah Schöngart et al. have used models of wealth groups and emissions data to discover the extent to which affluent groups become responsible for emissions, with "the wealthiest 10% [contributing] 6.5 times more," and "the top 1% and 0.1% contributing 20x and 76x more..."<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Honing in on the United States specifically, there are expectedly similar conclusions, as macroeconomist Fredrick N. G. Andersson states "income inequality is correlated with carbon emissions," admittedly with a weaker correlation level.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Regardless, these two articles are only a small portion of the many findings agreeing with higher-income groups having responsibility for the bulk of carbon emissions. As this idea is seen across nearly every nation, it would be reasonable to believe that smaller region levels such as San Diego would have similar results – though it would be interesting to see if there is such a disproportionate gap as with other regions.

Prior analysis utilizing appropriate datasets appear to reinforce the same idea. French economist Lucas Chancel conducted an analysis between income inequality and environmental impact datasets, revealing that the top 10% of earners in the United States and China emit greater tons of CO2 than the other 90% of earners in their respective countries, further highlighting this severe inequality emissions gap.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) Data analysis performed by ourworldindata.org into a chart also shows that high-income and upper-middle-income nations produce more than 30 billion tons of carbon dioxide combined – essentially 10x more than lower-income nations.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)

Many researchers have called to attention other factors surrounding income levels affecting carbon emissions; Amy Richmond and Robert Kauffman determined that energy prices provide a more direct measure of increases in emissions than strictly income per capita. They found that “real oil prices have a statistically measurable effect on per-capita energy use and carbon emissions,” as “higher energy prices reduce energy use… because both firms and households can substitute capital or labor."<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6) Ultimately, this concept supports the idea that an increase in income is associated with an increase in carbon emissions, as individuals with higher incomes can afford more energy, leading to producing more emissions than those with lower incomes. Richmond and Kauffman’s findings prove a need to account for energy prices in our analysis; by only measuring carbon emissions for the year 2022 in San Diego County, we limit the amount of fluctuations in energy prices within our data.

It is important to note that much of the research and data surrounding carbon emissions focuses on limiting its overall production, determining the most efficient way the total amount of emissions can be reduced to limit the growth of climate change. For instance, Chancel suggests that “governments [should] develop better data on individual emissions to monitor progress towards sustainable lifestyles,” while Richmond and Kauffman propose that “raising real energy prices may be a more effective means for reducing energy use.” While our project is limited to investigating the relationship between income levels and carbon emissions, our data will fit into the larger scope of evaluating how outside factors affect energy consumption, and in turn, contribute to climate change.

#### 2. Real Dataset

The dataset used in this project is publicly available at:  
https://data.openei.org/submissions/6219  

It provides census tract–level data for all of California from the **Low-Income Energy Affordability Data (LEAD) Tool – 2022 Update**. The dataset includes detailed measurements of income and residential energy expenditures. The key variables relevant to this project include:

- **AMI150**: Area Median Income grouping  
- **HINCP*UNITS**: Sum of average annual household income multiplied by number of units (used to compute total tract income and average income per household)
- **UNITS**: Number of households  
- **ELEP*UNITS**: Total electricity expenditures  
- **GASP*UNITS**: Total natural gas expenditures  
- **FULP*UNITS**: Total other fuel expenditures  

While carbon emissions are not explicitly included in the dataset, we estimate residential carbon emissions by converting these energy expenditures into physical energy use and then applying official emissions factors specific to San Diego in 2022.

Specifically, for each San Diego census tract:

1. We aggregate expenditure totals and household counts across income bins to produce one observation per tract.
2. We convert expenditures into estimated energy use using 2022 SDG&E residential price scalars.
3. We convert estimated energy use into CO₂ emissions using EPA emissions factors.
4. We normalize emissions by the number of households to obtain **estimated kg CO₂ per household per year**.

Formally:

kWh_t = ElecExp_t / P_elec,SD2022  

Therms_t = GasExp_t / P_gas,SD2022  

CO2e_t = (kWh_t × EF_elec,2022) + (Therms_t × EF_gas)  

CO2ePerHH_t = CO2e_t / UNITS_t  

Where:

- **P_elec,SD2022 = $0.37637/kWh**  
  (SDG&E TOU-DR1 Residential Total Electricity Cost, June 1, 2022)  
  Source: https://www.sdge.com/sites/default/files/2022.06.01%20CEA%20SDGE%20JRC_Final_0.pdf  

- **EF_elec,2022 = 499.3 lb CO₂e per MWh**  
  (EPA eGRID2022 WECC California CAMX region emission rate; converted to kg/kWh in code)  
  Source: https://www.epa.gov/system/files/documents/2024-01/egrid2022_summary_tables.pdf  

- **P_gas,SD2022 ≈ \$2.30 per therm**  
  (Derived from SDG&E 2022 rates fact sheet: $124 for 54 therms)  
  Source: https://autl.assembly.ca.gov/sites/autl.assembly.ca.gov/files/SDG%26E%20Electric%20and%20Gas%20Rates%20Fact%20Sheet%202022%20-%203-29-22.pdf  

- **EF_gas = 5.306 kg CO₂ per therm**  
  (EPA stationary combustion factor: 53.06 kg CO₂/mmBtu; 1 therm = 0.10 mmBtu per EIA)  
  Sources:  
  https://www.epa.gov/system/files/documents/2025-01/ghg-emission-factors-hub-2025.pdf  
  https://www.eia.gov/tools/faqs/faq.php?id=45&t=8  

This approach produces an estimated tract-level measure of residential carbon emissions per household, consistent with our research question examining the relationship between income and residential carbon emissions in San Diego County.

The scatter plot shows a clear positive linear relationship between average household income and estimated residential CO₂e emissions per household. As income increases, estimated emissions per household tend to increase as well. The fitted regression line indicates a strong upward trend, suggesting that higher-income tracts generally have higher residential carbon footprints.

Most tracts cluster between roughly \\$80,000 and \\$150,000 in average income and between about 1,000 and 2,500 kg CO₂e per household, with variability widening at higher income levels (mild heteroskedasticity). A smaller number of high-income tracts (above \\$250,000) exhibit emissions exceeding 3,000 kg and in some cases approaching 4,500 kg, contributing to a right tail in both income and emissions. However, there is also an outlier data point that shows around \\$380,000 income tract (the highest in our dataset), but only exhibits emissions around 3,000 kg; this is most likely noise, as this data point is the only one that is notably distant from the overall positive trend and falls substantially below the regression line relative to other high-income tracts.

Overall, the pattern suggests a statistically meaningful and approximately linear association, supporting the hypothesis that household income is a strong predictor of residential carbon emissions in this dataset.

```python
plt.figure(figsize=(8,5))

sns.regplot(
    data=df_sd_clean,
    x="log_income",
    y="log_co2e",
    line_kws={"color":"red"}
)

plt.title("Log(Income) vs Log(Estimated Residential Emissions)\nSan Diego Census Tracts (2022)")
plt.xlabel("Log Average Household Income")
plt.ylabel("Log Estimated CO₂e per Household")
plt.tight_layout()
plt.show()
```

The log–log scatter plot of average household income and estimated residential CO₂e emissions per household reveals a strong and approximately linear positive relationship across San Diego census tracts (2022). After log transformation, the association appears more structurally linear compared to the raw scale, with the previously noticeable highest-income data point appearing less visually deviant from the overall trend. No single observation exerts disproportionate visual influence in the transformed graph.

In log space, estimated emissions range roughly from 6.5 to 8.5 (log kg CO₂e per household), while log average household income spans approximately 10 to 13. The majority of tracts cluster between 7 and 8 log emissions and 11 to 12 log income. While this central concentration is partly a mathematical consequence of logarithmic compression of the original right-skewed distributions, it also reflects the underlying distribution of middle-income census tracts in the dataset.

Variability appears lowest among middle-income tracts and increases toward both the lower and upper ends of the income distribution. This pattern suggests that while emissions scale proportionally with income on average, greater heterogeneity exists among the lowest- and highest-income tracts. Importantly, compared to the raw-scale model, variance appears more symmetric and less systematically increasing with income, indicating that the log transformation improves homoscedasticity. This is on top of the fact that the variance at each ends of the transformed graph are generally also smaller compared to the raw graph.

Overall, the log–log specification supports a proportional interpretation of the relationship: percentage increases in income are associated with percentage increases in residential emissions. The improved linearity, reduced visual presence of the extreme observation, and more stable variance structure suggest that the log–log model may provide a more statistically appropriate and interpretable framework for regression analysis in this context, although both specifications yield substantively similar conclusions.

#### Income-Grouped Differences in Residential Emissions (Quartiles)

Scatter plots show an overall positive association between household income and estimated CO₂ emissions per household, but they may obscure how emissions vary across the income distribution. To examine this more closely, we group census tracts (small geographic areas used by the U.S. Census Bureau to represent neighborhoods) into income quartiles and compare the distribution of emissions within each group. This allows us to evaluate whether higher-income areas consistently produce higher emissions or if the pattern is driven by a small number of extreme observations.

```python
# create income quartiles (4 groups). Use qcut for balanced groups
# if qcut fails due to ties, get rid of duplicates.
df_sd_clean["income_quartile"] = pd.qcut(
    df_sd_clean["income_per_hh"],
    q=4,
    labels=["Q1 (lowest)", "Q2", "Q3", "Q4 (highest)"],
    duplicates="drop"
)

# get the bin boundaries used by qcut
_, bins = pd.qcut(
    df_sd_clean["income_per_hh"],
    q=4,
    retbins=True,
    duplicates="drop"
)

# summary table with mean and median emissions by income quartile
quartile_summary = (
    df_sd_clean.groupby("income_quartile")["co2e_per_hh_kg"]
    .agg(["count", "mean", "median", "std"])
    .reset_index()
)

# add ranges of income that qcut split into
ranges = [
    f"${int(bins[i]):,}–${int(bins[i+1]):,}"
    for i in range(len(bins)-1)
]

quartile_summary["income_range"] = ranges

display(quartile_summary)

# visualize boxplot of emissions by income quartile
plt.figure(figsize=(9,5))
sns.boxplot(data=df_sd_clean, x="income_quartile", y="co2e_per_hh_kg")
plt.title("Estimated CO₂e per Household by Income Quartile\nSan Diego Census Tracts (2022)")
plt.xlabel("Income Quartile (by average household income)")
plt.ylabel("Estimated CO₂e per Household (kg/year)")
plt.tight_layout()
plt.show()
```

The median emissions increase steadily across income quartiles, suggesting that the relationship between income and residential emissions is not driven solely by a small number of extreme outliers. Instead, the upward shift in the median across groups indicates a broader pattern in which higher income census tracts tend to produce more emissions per household. Additionally, the standard deviation increases across higher income quartiles, indicating greater variability in emissions among wealthier areas. This may reflect the wider range of housing types, energy consumption patterns, and lifestyle choices available to higher income households. However, there is still some overlap between the distributions of emissions across quartiles, suggesting that factors beyond income also influence residential emissions.

#### Limitations

#### ANOVA and Tukey HSD Analysis

A one-way ANOVA was conducted to test whether estimated residential carbon emissions differ across Area Median Income (AMI150) categories. Because group variances were unequal, Welch’s ANOVA was applied. Below shows the code and the table of the result that highlights the important output values.

The results indicate a statistically significant difference in mean emissions across the multiple income groups with F(5, 54.89) = 17.58, and most notably the p value way below 0.001. This finding suggests that residential carbon emissions vary systematically across income categories in San Diego census tracts. Examination of group means shows that higher-income tracts generally exhibit higher average emissions per household, supporting the hypothesis that income level is positively associated with residential carbon emissions.

```python
import statsmodels.api as sm

anova_result = sm.stats.anova_oneway(
    data=df_sd_clean["co2e_per_hh_kg"],
    groups=df_sd_clean["AMI150"]
)
```

| Test                  | F-statistic | df (num, denom) | p-value    | Notes                                   |
|-----------------------|------------|----------------|-----------|----------------------------------------|
| Welch ANOVA           | 17.58      | 5, 54.89       | 2.33e-10  | Significant differences among groups; higher-income tracts have higher emissions |

Post-hoc Tukey HSD tests were conducted following the significant Welch ANOVA to examine which AMI150 income groups differed in residential carbon emissions. Several statistically significant pairwise differences were observed.
```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Ensure DataFrame structure
tukey_df = pd.DataFrame(
    data=tukey_result.summary().data[1:],
    columns=tukey_result.summary().data[0]
)

# Ensure numeric columns are numeric
for col in ["meandiff", "lower", "upper", "p-adj"]:
    tukey_df[col] = pd.to_numeric(tukey_df[col])

# Ordered AMI groups
group_order = ["0-30%", "30-60%", "60-80%", "80-100%", "100-150%", "150%+"]
order_map = {g: i for i, g in enumerate(group_order)}

# Normalize labels
tukey_df["group1"] = tukey_df["group1"].str.replace("–", "-", regex=False)
tukey_df["group2"] = tukey_df["group2"].str.replace("–", "-", regex=False)

# Sort to always lower-order group first
def canonicalize_row(row):
    g1, g2 = row["group1"], row["group2"]
    md, lo, hi = row["meandiff"], row["lower"], row["upper"]
    
    if order_map[g1] <= order_map[g2]:
        return pd.Series({
            "group1_clean": g1,
            "group2_clean": g2,
            "meandiff_clean": md,
            "lower_clean": lo,
            "upper_clean": hi
        })
    else:
        return pd.Series({
            "group1_clean": g2,
            "group2_clean": g1,
            "meandiff_clean": -md,
            "lower_clean": -hi,
            "upper_clean": -lo
        })

cleaned = tukey_df.apply(canonicalize_row, axis=1)
tukey_df = pd.concat([tukey_df, cleaned], axis=1)

# Ordering variables
tukey_df["g1_order"] = tukey_df["group1_clean"].map(order_map)
tukey_df["g2_order"] = tukey_df["group2_clean"].map(order_map)
tukey_df["gap"] = tukey_df["g2_order"] - tukey_df["g1_order"]

# Clean label
tukey_df["comparison"] = tukey_df["group1_clean"] + " vs " + tukey_df["group2_clean"]

# Sort: adjacent comparisons first, then by first group
tukey_df = tukey_df.sort_values(["gap", "g1_order", "g2_order"]).reset_index(drop=True)

# Plot
fig, ax = plt.subplots(figsize=(10.5, 7.5))
y = np.arange(len(tukey_df))

for i, row in tukey_df.iterrows():
    sig = bool(row["reject"])
    color = "tab:blue" if sig else "0.75"
    lw = 2.2 if sig else 1.8
    
    ax.plot(
        [row["lower_clean"], row["upper_clean"]],
        [i, i],
        color=color,
        linewidth=lw,
        solid_capstyle="round"
    )
    ax.scatter(
        row["meandiff_clean"],
        i,
        color=color,
        s=55 if sig else 40,
        zorder=3
    )

# Reference line at zero
ax.axvline(0, color="black", linestyle="--", linewidth=1)

# Signficance asterisk in label
tukey_df["comparison"] = tukey_df.apply(
    lambda r: f"{r['group1_clean']} vs {r['group2_clean']}" + (" *" if r["reject"] else ""),
    axis=1
)

# Labels and formatting
ax.set_yticks(y)
ax.set_yticklabels(tukey_df["comparison"])
ax.invert_yaxis()

ax.set_xlabel("Mean difference in CO₂e per household (kg)")
ax.set_title("Tukey HSD pairwise comparisons across AMI groups", pad=12)

# Styling and saving
ax.set_facecolor("white")
fig.patch.set_facecolor("white")
ax.grid(axis="x", color="0.9", linewidth=0.8)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.savefig("results/tukey_hsd_comparison.png", dpi=300, bbox_inches="tight")
plt.close(fig)
```

Points show estimated pairwise mean differences in household CO₂e emissions and bars indicate 95% confidence intervals; positive differences mean the second group emitted more than the first. Blue intervals mark statistically significant comparisons (α = 0.05). The results indicate that lower-income tracts, especially the 0–30% AMI group, had significantly lower emissions than several middle- and upper-middle-income groups, while comparisons involving the 150%+ AMI group were not significant due to wide uncertainty. Overall, the pattern supports a systematic increase in residential carbon emissions with income.

#### Welch's t-Test of Low vs High Income Tracts

To complement the ANOVA and Tukey HSD analyses, we conducted a Welch’s t-test to compare residential CO₂e emissions between low-income (0–30% AMI) and high-income (100–150% + 150%+ AMI) census tracts. This test was chosen because the two groups have unequal variances and sample sizes. Below shows the code and the summary result table.

The results indicated a statistically significant difference in emissions between the two groups: t(approx. df) = 6.06, which shows that high-income tracts emit more CO₂e per household on average and p value way below 0.001, meaning it is statistically significant.

```python
from scipy import stats

# Define low vs high income 
low_income = df_sd_clean[df_sd_clean['AMI150'] == '0-30%']['co2e_per_hh_kg']
high_income = df_sd_clean[df_sd_clean['AMI150'].isin(['100-150%', '150%+'])]['co2e_per_hh_kg']

t_val, p_val = stats.ttest_ind(high_income, low_income, equal_var=False)
```

| Test                 | t-statistic | p-value      | Comparison                  |
|----------------------|------------|-------------|----------------------------|
| Welch t-test         | 6.06       | 3.05e-08    | High vs Low Income Tracts  |

The corresponding effect size was Cohen’s d = 0.84, which is considered a large effect, indicating that the difference is not only statistically significant but also practically meaningful. Below shows the code and the summary result table.

In other words, high-income households in San Diego County produce substantially more residential carbon emissions than low-income households, consistent with the trends observed in both the regression and Tukey HSD analyses.

```python
import numpy as np

# Means
mean_high = high_income.mean()
mean_low = low_income.mean()

# Standard deviations
std_high = high_income.std()
std_low = low_income.std()

# Sample sizes
n_high = len(high_income)
n_low = len(low_income)

# Cohen's d (Welch-adjusted)
pooled_std = np.sqrt((std_high**2 + std_low**2) / 2)
cohens_d = (mean_high - mean_low) / pooled_std
```

In conclusion, households within the San Diego 2022 census tracts that produce a higher income have shown a convincing correlation with higher estimated carbon emissions, aligning with our original hypothesis. Initial analysis of the dataset shows a distribution that is right-skewed for both household income and carbon emissions; and after comparing and visualizing using scatterplots and boxplots, households that produce a larger amount of income produce a considerable amount more carbon compared to lower-income households. Further statistical analysis including ANOVA, Tukey HSD, and Welch's t-test highlight a practically significant difference in addition to the already statistically significant difference seen between higher and lower income tracts. To reiterate, analysis of the dataset highlights a significant correlation between household income within San Diego census tracts and carbon emissions.

While perhaps not adding anything groundbreaking to the existing research on this topic, finding similar results on a smaller, local area that may not have been examined before may help build further consensus. However, it may be of interest to compare the extent to which higher-income regions produce carbon compared to lower-income regions from a local to a wider scale; referring back to ourworldindata.org's data analysis highlighting how high-/upper-middle-income nations produce more than 10x carbon dioxide than lower-income nations, the difference in terms of San Diego households does not seem as exaggerated. Of course households differ in carbon emissions compared to entire nations, but this could prove useful in the future regardless.

Moving towards the bigger picture, it should be apparent that some form of change is needed in order to prevent the excessive usage of electricity and fuels---and thus the excessive emissions of carbon---from higher-income households. Of course, this is easier said than done; considering said households may likely be used to their carbon-heavy lifestyles, they may be more resistant to change, especially when it makes their lives more inconvenient as a result. It may then be up to the government to make the transition process easier, whether it be trying to utilize eco-friendly sources of energy in lieu of harmful fossil fuels or promoting benefits for changing to a less carbon emissions-heavy lifestyle.

Although our results consistently show agreement with our hypothesis, along with some reassurance that selection bias may be minimized due to the millions of observations within our data, there may still exist certain limitations. As we use a proxy for energy expenditure to account for emissions, this introduces confounds such as intra-year fluctuations in energy prices, which may have partly affected our findings. This would make it difficult to expand the scope of this project past a one-year period, as these fluctuations prevent accurate tracking of yearly changes in carbon emissions, complicating long-term comparative analysis.

Although our models controlled for several confounding variables, including race, education level, and housing type, some factors remained unaccounted for. For example, the average age of homes in a given neighborhood may be a factor, as tracts with older, poorly insulated houses may require more energy for heating than tracts with newer, better insulated homes. Household size is another potential confounder, as larger households typically consume more energy than smaller households. At the same time, aggregating data at the census tract level prevents capturing income distribution within each tract. Gentrification in neighborhoods across San Diego increases income inequality within tracts, potentially affecting our income classifications and, in turn, model accuracy. However, the strong relationship between income distribution and carbon emissions suggests that the influence of these unmeasured factors was limited.

To expand the scope of this project, future researchers could expand this frame of analysis to additional cities. Comparative studies at the state or national level may reveal where the most significant disparities in carbon emissions by income exist, and identify common factors among cities with the largest gaps.
